### 문서 로딩

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("E:/work/MS_project_2/data/test.md", encoding="utf-8")
documents = loader.load()
```

### 청크분할(langchain사용)

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,    # 한 청크의 최대 길이
    chunk_overlap=50,  # 겹치는 길이 (문맥 유지를 위해)
)

docs_split = text_splitter.split_documents(documents)

# 결과 확인
print(f"총 {len(docs_split)}개의 문서 청크 생성됨.")
print(docs_split[0].page_content)  # 첫 번째 청크 예시 출력
```

```python
print(docs_split[6].page_content)
```

```python
from openai import AzureOpenAI
from azure.core.credentials import AzureKeyCredential

# ✅ 필요한 정보
api_key = ""
endpoint = ""
api_version = "2024-02-15-preview"
deployment = "text-embedding-3-small"

# ✅ AzureOpenAI 클라이언트 생성
client = AzureOpenAI(
    api_version=api_version,
    azure_endpoint=endpoint,
    api_key=api_key
)

# ✅ 문서 청크 리스트 만들기 (docs_split에 이미 존재한다고 가정)
texts = [doc.page_content for doc in docs_split]

# ✅ 임베딩 요청
response = client.embeddings.create(
    input=texts,
    model=deployment
)

# ✅ 벡터 출력 예시
for item in response.data:
    print(f"청크 {item.index} → 벡터 길이: {len(item.embedding)}")
```

### 임베딩 객체

```python
from langchain_openai import AzureOpenAIEmbeddings
import os
os.environ.pop("OPENAI_API_BASE", None)
os.environ.pop("BASE_URL", None)
embedding = AzureOpenAIEmbeddings(
    api_key=api_key,
    azure_endpoint=endpoint,
    model=deployment,  # 배포 이름 확인 필수
    openai_api_version=api_version
)
```

```python
import faiss
import numpy as np
from langchain.schema import Document
from langchain.vectorstores.faiss import FAISS

# 1. 이미 생성된 임베딩 벡터와 문서 청크 준비
#    texts: 각 문서 청크의 텍스트 (문자열 리스트)
#    docs_split: 각 문서 청크가 담긴 Document 객체 리스트 (각 Document의 page_content는 문자열이어야 함)
#    vectors: response.data에서 추출한 임베딩 벡터 리스트 (각 벡터는 list[float] 타입)

# 예시)
# texts = [doc.page_content for doc in docs_split]
# vectors = [item.embedding for item in response.data]

# 2. 임베딩 벡터들을 NumPy 배열로 변환 (FAISS는 float32 배열을 사용)
embeddings_np = np.array(vectors).astype("float32")

# 3. FAISS 인덱스 생성 (차원은 임베딩 벡터의 길이와 동일)
dimension = embeddings_np.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings_np)


```

```python
# 4. 문서 매핑 생성: 인덱스 번호를 문서로 매핑하는 dict 생성
#    LangChain의 FAISS 클래스는 docstore: dict[str, Document]와
#    index_to_docstore_id: dict[int, str]를 필요로 합니다.
from langchain.docstore.in_memory import InMemoryDocstore

docstore = InMemoryDocstore({str(i): doc for i, doc in enumerate(docs_split)})
index_to_docstore_id = {i: str(i) for i in range(len(docs_split))}


```

```python
# 5. FAISS 벡터스토어 생성
#    embedding_function은 쿼리 임베딩 계산에 사용됩니다.
#    여기서는 AzureOpenAIEmbeddings 객체를 사용하면 됩니다.
vectorstore = FAISS(
    embedding_function=embedding,  # 쿼리 텍스트에 대한 임베딩 계산에 사용 (예: AzureOpenAIEmbeddings)
    index=index,
    docstore=docstore,
    index_to_docstore_id=index_to_docstore_id
)

```

```python
# 6. 검색 테스트 (예: "신청 자격 요건"에 대한 유사 문서 검색)
query = "임대조건 1순위는?"
results = vectorstore.similarity_search(query, k=3)

for i, doc in enumerate(results):
    print(f"\n🔍 Top {i+1} 결과:\n{doc.page_content}")
```

### LLM 연동

```python
import requests
import re

# Azure OpenAI GPT-4o 설정
llm_endpoint = ""
llm_api_key = ""

def request_gpt(prompt: str) -> str:
    """
    GPT-4o에 프롬프트를 보내고 응답 받기
    """
    headers = {
        'Content-Type': 'application/json',
        'api-key': llm_api_key
    }

    body = {
        "messages": [
            {
                "role": "system",
                "content": "너는 친절하고 정확한 AI 도우미야. 사용자 질문에 문서 기반으로 답해줘."
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        "temperature": 0.7,
        "top_p": 0.95,
        "max_tokens": 800
    }

    response = requests.post(llm_endpoint, headers=headers, json=body)

    if response.status_code == 200:
        response_json = response.json()
        message = response_json['choices'][0]['message']
        content = message['content']
        # [doc1] 같은 포맷이 있을 경우 사람이 읽기 쉽게 변환
        content = re.sub(r'\[doc(\d+)\]', r'[참조 \1]', content)
        return content
    else:
        print("❌ 요청 실패:", response.status_code, response.text)
        return "⚠️ 오류가 발생했습니다."
```

```python
def generate_answer_with_rag(query: str, vectorstore, top_k: int = 3) -> str:
    """
    FAISS에서 관련 문서를 검색한 뒤, 그 내용을 기반으로 GPT-4o에 질문하는 함수
    """
    results = vectorstore.similarity_search(query, k=top_k)

    context = "\n\n".join([f"[doc{i+1}]\n{doc.page_content}" for i, doc in enumerate(results)])

    prompt = f"""다음은 사용자가 질문한 내용과 관련된 문서 내용이야. 이 문서를 참고해서 질문에 대해 정확하고 구체적으로 답변해줘.

[사용자 질문]
{query}

[참고 문서]
{context}

답변:"""

    return request_gpt(prompt)
```

```python
query = "임대조건 1순위는?"

# RAG 없이 일반 GPT 사용 (배경지식 기반)
basic_answer = request_gpt(query)
print("📄 GPT 일반 답변:\n", basic_answer)

# RAG 기반 답변 생성 (문서 + LLM)
rag_answer = generate_answer_with_rag(query, vectorstore)
print("📚 RAG 기반 GPT 답변:\n", rag_answer)
```
